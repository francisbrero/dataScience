```python
from langchain import OpenAI, SQLDatabase, SQLDatabaseChain
from langchain.chat_models import ChatOpenAI
import os
from dotenv import find_dotenv, load_dotenv

# Load environment variables from .env file
load_dotenv(find_dotenv())

# Set API credentials as variables => Please make sure you have a .env file with the following variables
OPENAI_API_KEY = os.environ["OPENAI_API_KEY"]
# configure the postgres connection
pg_user = os.environ["pg_user"]
pg_emback_pwd = os.environ["pg_emback_pwd"]
pg_host = os.environ["pg_host"]
pg_db = os.environ["pg_db"]
pg_schema = os.environ["pg_schema"]

# Connect to the database
db = SQLDatabase.from_uri(
    f"postgresql+psycopg2://{pg_host}:5432/{pg_db}?sslmode=require&user={pg_user}&password={pg_emback_pwd}",
    schema=f"{pg_schema}",
    include_tables=["account_computations_custom","activities_persons","activities_accounts"],
    )

# setup llm
llm = ChatOpenAI(
    temperature=0, 
    verbose=True, 
    openai_api_key=OPENAI_API_KEY,
    model_name='gpt-3.5-turbo', # uses fewer tokens than davinci
    max_tokens=1000
    )
```
